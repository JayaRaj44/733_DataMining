```python
import pandas as pd
import numpy as np
data = pd.read_csv('C:/Users/rajgo/Downloads/Baltimore911.csv')
# Total number of rows and columns
num_rows = data.shape[0]
num_columns = data.shape[1]

# Create an empty list to store rows of profile information
profile_rows = []

# Iterate through each column in the dataset
for column_name in data.columns:
    column_data = data[column_name]
    data_type = str(column_data.dtypes)

    # Check if the column is numeric or not
    if np.issubdtype(column_data, np.number):
        min_val = column_data.min()
        max_val = column_data.max()
        mean_val = column_data.mean()
        median_val = column_data.median()
        std_dev = column_data.std()
        unique_values = None
        top_3_values = None
        missing_percentage = (num_rows - column_data.count()) / num_rows * 100
    else:
        min_val = None
        max_val = None
        mean_val = None
        median_val = None
        std_dev = None
        unique_values = column_data.nunique()
        top_3_values = column_data.value_counts().head(3)
        missing_percentage = (column_data.isnull().sum() / num_rows) * 100

    # Append the profile information for the column as a dictionary
    profile_rows.append({
        'Column Name': column_name,
        'Data Type': data_type,
        'Min': min_val,
        'Max': max_val,
        'Mean': mean_val,
        'Median': median_val,
        'Std Dev': std_dev,
        'Unique Values': unique_values,
        'Top 3 Values': top_3_values,
        'Missing (%)': missing_percentage
    })

# Create a DataFrame from the list of profile information
profile_data = pd.DataFrame(profile_rows)

# Display the dataset profile
print(profile_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame with a 'Date' column representing the timestamp of each crime incident
# Replace 'your_dataset.csv' with the path to your actual dataset
df = pd.read_csv('C:/Users/rajgo/Downloads/Baltimore911.csv')

# Convert the 'Date' column to datetime
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])

# Extract year, month, and day-of-week from the 'Date' column
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.dayofweek  # Monday is 0 and Sunday is 6

# Group by year, month, and day-of-week and count the number of incidents
yearly_counts = df.groupby('Year').size()
monthly_counts = df.groupby('Month').size()
day_of_week_counts = df.groupby('DayOfWeek').size()

# Plotting
# Day-of-Week
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(12, 6))
day_of_week_counts.plot(kind='bar', color='yellow')
plt.title('Day-of-Week Crime Incidence')
plt.xlabel('Day of Week')
plt.xticks(ticks=range(7), labels=days_of_week, rotation=45)
plt.ylabel('Number of Incidents')
plt.show()

# Year-to-Year
plt.figure(figsize=(12, 6))
yearly_counts.plot(kind='bar', color='purple')
plt.title('Year-to-Year Crime Incidence')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.show()

# Monthly
plt.figure(figsize=(12, 6))
monthly_counts.plot(kind='bar', color='red')
plt.title('Monthly Crime Incidence')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (replace 'your_dataset.csv' with your actual file)
data = pd.read_csv('C:/Users/rajgo/Downloads/Baltimore911.csv')

# Convert the 'Date' column to datetime
data['CrimeDate'] = pd.to_datetime(data['CrimeDate'])

# Extract year, month, and day of the week
data['Year'] = data['CrimeDate'].dt.year
data['Month'] = data['CrimeDate'].dt.month
data['DayOfWeek'] = data['CrimeDate'].dt.day_name()

# Create subplots for year-to-year, monthly, and day-of-week crime incidence
plt.figure(figsize=(15, 5))

# Year-to-Year Crime Incidence
plt.subplot(131)
sns.countplot(data=data, x='Year', palette='viridis')
plt.title('Year-to-Year Crime Incidence')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')

# Monthly Crime Incidence
plt.subplot(132)
sns.countplot(data=data, x='Month', palette='viridis')
plt.title('Monthly Crime Incidence')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')

# Day-of-Week Crime Incidence
plt.subplot(133)
sns.countplot(data=data, x='DayOfWeek', palette='viridis', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Day-of-Week Crime Incidence')
plt.xlabel('Day of the Week')

plt.tight_layout()
plt.show()
```

```python
import dash
from dash import html, dcc, Input, Output
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
import dash_leaflet as dl

# Read the CSV file and specify data types for columns (if needed)
df = pd.read_csv('C:/Users/rajgo/Downloads/Baltimore911.csv')

# Convert the 'CrimeDate' column to datetime
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])

# Extract year, month, and day-of-week from the 'CrimeDate' column
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.dayofweek  # Monday is 0 and Sunday is 6

# Create the Dash web application
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Crime Exploration Dashboard"),

    # Dropdown for selecting location
    dcc.Dropdown(
        id='location-dropdown',
        options=[{'label': location, 'value': location} for location in df['Location'].unique()],
        value=df['Location'].unique()[0],
        multi=False,
        style={'width': '50%'}
    ),

    # Spatial map for crime locations
    dl.Map(id="crime-location-map", style={'width': '100%', 'height': '500px'}, center=[39.29, -76.61], zoom=11,
           children=[
               dl.TileLayer(),
               dl.MarkerCluster(id="marker-cluster", children=[]),
           ]),

    # Temporal pattern chart (Year-to-Year)
    dcc.Graph(id='temporal-pattern-chart'),

    # Bar chart for crime type distribution
    dcc.Graph(id='crime-type-distribution-chart')
])

# Define callback to update charts and map based on user input
@app.callback(
    [Output('temporal-pattern-chart', 'figure'),
     Output('crime-type-distribution-chart', 'figure'),
     Output("marker-cluster", "children")],
    [Input('location-dropdown', 'value')]
)
def update_charts(selected_location):
    # Filter data based on selected location
    filtered_df = df[df['Location'] == selected_location]

    # Temporal pattern chart (Year-to-Year)
    yearly_counts = filtered_df.groupby('Year').size().reset_index(name='Incidents')
    temporal_pattern_chart = px.line(yearly_counts, x='Year', y='Incidents', title='Year-to-Year Temporal Pattern')

    # Crime type distribution chart
    crime_type_counts = filtered_df['Description'].value_counts().reset_index(name='Incidents')
    crime_type_distribution_chart = px.bar(crime_type_counts, x='Description', y='Incidents', title='Crime Type Distribution')

    # Spatial map for crime locations
    markers = [
        dl.Marker(position=[row['Latitude'], row['Longitude']], children=[
            dl.Tooltip(row['Description'])
        ])
        for _, row in filtered_df.iterrows()
    ]

    return temporal_pattern_chart, crime_type_distribution_chart, markers


if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
